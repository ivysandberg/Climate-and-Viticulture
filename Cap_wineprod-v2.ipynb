```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df_onehot = pd.get_dummies(df, columns=['Varietal'], prefix = ['Varietal'])
df_onehot.head()
```

```python
df_onehot.shape
```

```python
df_onehot = df_onehot.drop(['Prod_TotalValue', 'Bearing', 'Non-Bearing', 'AcreageTotal'], axis=1)
```

```python
# assign features
features=df_onehot.loc[:, df_onehot.columns != 'Production_tons']
# remove year
# features = features.drop(['Year'], axis=1)
features.info()
```

```python
label=df_onehot['Production_tons']

# Training Data
from sklearn.model_selection import train_test_split
train_data, test_data, train_labels, test_labels = train_test_split(features, label, test_size=0.3, random_state=42)
```

```python
# Model Comparison

from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor

def compare_models(model):
    model_name = model.__class__.__name__
    fit=model.fit(train_data,train_labels)
    y_pred=fit.predict(test_data)
    r2=r2_score(test_labels,y_pred)
    return([model_name,r2])

models = [
    GradientBoostingRegressor(n_estimators=200, max_depth=3, random_state=0),
     RandomForestRegressor(n_estimators=200, max_depth=3, random_state=0),
   DecisionTreeRegressor()
]

model_train=list(map(compare_models,models))
print(*model_train, sep = "\n")
```

```python
# df_onehot = df_onehot.drop(['Year'], axis=1)
df_onehot.head()
```

```python
test_df=pd.DataFrame(test_data,columns=df_onehot.loc[:, df_onehot.columns != 'Production_tons'].columns) 

# using stack function to return a reshaped DataFrame by pivoting the columns of the current dataframe
var=test_df[[col for col in test_df.columns if 'Varietal' in col]].stack()[test_df[[col for col in test_df.columns if 'Varietal' in col]].stack()>0]
varlist=list(pd.DataFrame(var).index.get_level_values(1))
varieties=[i.split("_")[1] for i in varlist]
```

```python
test_df.drop([col for col in test_df.columns if 'Varietal' in col],axis=1,inplace=True)
test_df['Varietal']=varieties
test_df.head()
```

```python
# from sklearn import tree
# clf=DecisionTreeRegressor()
# model=clf.fit(train_data,train_labels)
# tree.plot_tree(model)

clf=GradientBoostingRegressor()
model=clf.fit(train_data,train_labels)

# coefficient of determination R^2 of the prediction
print("Accuracy score: ", clf.score(test_data, test_labels))

test_df["predicted"]= model.predict(test_data)
test_df["actual"]=pd.DataFrame(test_labels)["Production_tons"].tolist()
test_group=test_df.groupby("Varietal")
```

```python
from sklearn.model_selection import cross_val_score

test_group.apply(lambda x: cross_val_score(clf, features, label, cv=10))
```

```python
fig, ax = plt.subplots() 

ax.scatter(test_df["actual"], test_df["predicted"],edgecolors=(0, 0, 0))

ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual vs Predicted")
plt.show()
```

```python
varimp= {'imp':model.feature_importances_,'names':df_onehot.columns[df_onehot.columns!="Production_tons"]}

#7 most important factors that affect production 
a4_dims = (16.7, 8.27)

fig, ax = plt.subplots(figsize=a4_dims)
df1=pd.DataFrame.from_dict(varimp)
df1.sort_values(ascending=False,by=["imp"],inplace=True)
df1=df1.dropna()
df1=df1.nlargest(15, 'imp')
sns.barplot(x="imp",y="names",palette="vlag",data=df1,orient="h",ax=ax);
```
